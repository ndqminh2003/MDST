Let's compare Michigan to Ohio. Which state has the higher average outcome score across all years?

```python
michigan_data = df[df['STATE'] == 'MICHIGAN']
ohio_data = df[df['STATE'] == 'OHIO']
michigan_avg_score = michigan_data['YEAR'].mean().mean()
ohio_avg_score = ohio_data['YEAR'].mean().mean()
if michigan_avg_score > ohio_avg_score:
    print("Michigan has a higher average outcome score.")
elif michigan_avg_score < ohio_avg_score:
    print("Ohio has a higher average outcome score.")
else:
    print("Michigan and Ohio have the same average outcome score.")
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# X =
# y = 
X = df[['ENROLL','AVG_MATH_4_SCORE','AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_READING_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
# model = 
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.intercept_)
print(model.coef_)
model.score(X_test, y_test)
```
